```python
# W3schools tutorial Python MongoDB

dbName = 'pythonDatabase'
colName = 'w3SchoolTutorial'
mycol = myclient[dbName][colName]


def listDatabase(dbClient):
    # Return a list of your system's databases:
    print(dbClient.list_database_names())
    return dbClient.list_database_names()

def checkDatabase(dbClient, dbName):
    # Creating a Database if not exist.
    if dbName in dbClient.list_database_names():
        print(dbName, ": Database exists.")
        return True
    else :
    # Creates a new database 'mydatabase'
        print('Initiate dbclient.dbname object through: tmp = client.dbName ')
        return False


# Creating a Collection
def listCollection(dbClient, dbName):
    print(dbClient[dbName].list_collection_names())
    return dbClient[dbName].list_collection_names()

# collist = mydb.list_collection_names()
def checkCollection (dbClient, dbName, dbCollection):
    if dbCollection in dbClient[dbName].list_collection_names():
        print(dbCollection, ": Collection exists.")
        listCollection(dbClient, dbName)
        return True
    else :
        # Creates a new collection 'customers'
        print('Initiate dbclient.dbname.col object through: tmp = client.dbName.col ')
        return False



# Insert Into Collection
def insertOneIntoCol(dbClient, dbName, dbCollection, dictJ):
    x = dbClient[dbName][dbCollection].insert_one(dictJ)
    print('inserted_id:', x.inserted_id)
    return x.inserted_id


# Insert Multiple Documents
def insertManyIntoCol(dbClient, dbName, dbCollection, listDict):
    xn = dbClient[dbName][dbCollection].insert_many(listDict)
    #print list of the _id values of the inserted documents:
    print(xn.inserted_ids)
    xOut = []
    for x in xn.inserted_ids : xOut.append(x)
    return xOut


# Insert Multiple Documents, with Specified IDs
def insertManyIdIntoCol(dbClient, dbName, dbCollection, listDict):
    xnId = dbClient[dbName][dbCollection].insert_many(listDict)
    print(xnId.inserted_ids)
    #print list of the _id values of the inserted documents:
    for x in xnId : findQueryCol(dbClient, dbName, dbCollection, {'_id': x})
    xOut = []
    for x in xnId : xOut.append(x)
    return xOut



# Find by ID
def findById(dbClient, dbName, dbCollection, byID):
    print(dbClient[dbName][dbCollection].find({'_id' : byID}))
    return dbClient[dbName][dbCollection].find({'_id' : byID})

# Find One
def findFirstInCol(dbClient, dbName, dbCollection):
    print(dbClient[dbName][dbCollection].find_one())
    return dbClient[dbName][dbCollection].find_one()

# Find All
def findAllInCol(dbClient, dbName, dbCollection):
    xOut = []
    for x in dbClient[dbName][dbCollection].find():
        print(x)
        xOut.append(x)
    return xOut

# Limit the Result
def headInCol(dbClient, dbName, dbCollection):
    xOut = []
    for x in dbClient[dbName][dbCollection].find().limit(5):
        print(x)
        xOut.append(x)
    return xOut


# Return Only Some Fields
# The second parameter of the find() method is an object describing which fields 
# to include in the result.
# This parameter is optional, and if omitted, all fields will be included in the
# result.
# for x in mycol.find({},{ "_id": 0, "name": 1, "address": 1 }):
#     print(x)


def findQueryCol(dbClient, dbName, dbCollection, queryDict):
    xOut = []
    for x in dbClient[dbName][dbCollection].find(queryDict):
        print(x)
        xOut.append(x)
    return xOut

def filterQueryCol(dbClient, dbName, dbCollection, queryDict, filterQuery):
    xOut = []
    for x in dbClient[dbName][dbCollection].find(queryDict, filterQuery):
        print(x)

def findQueryColAscendSortBy(dbClient, dbName, dbCollection, queryDict, sortKey):
    # Sort the Result Ascending
    xOut = []
    for x in dbClient[dbName][dbCollection].find(queryDict).sort(sortKey):
        print(x)
        xOut.append(x)
    return xOut

def findQueryColDescendSortBy(dbClient, dbName, dbCollection, queryDict, sortKey):
    # Sort the Result Descending
    xOut = []
    for x in dbClient[dbName][dbCollection].find(queryDict).sort(sortKey, -1):
        print(x)
        xOut.append(x)
    return xOut

# for x in mycol.find({ "address": { "$regex": "^S" } }): print(x)
# for x in mycol.find(): print(x)

def updateOneInCol(dbClient, dbName, dbCollection, queryDict, UpdateDict):
    # Update collection. SetUpdateDict expects: {key: value}, {"$set": {key:value}}
    dbClient[dbName][dbCollection].update_one(queryDict, {"$set": UpdateDict})
    findQueryCol(dbClient, dbName, dbCollection, UpdateDict)

# Update Collection
# mycol.update_one({ "address": "Valley 345" }, 
                # { "$set": { "address": "Canyon 123" } })

def updateManyInColRegex(dbClient, dbName, dbCollection, queryRegexDict, UpdateDict):
    # SetUpdateDict expects: {key: {'$regex':value}}, {"$set": {key:value}}
    dbClient[dbName][dbCollection].update_many(queryRegexDict, {"$set": UpdateDict})
    findQueryCol(dbClient, dbName, dbCollection, UpdateDict)

# Update Many
# print(mycol.update_many({ "address": { "$regex": "^S" } }, 
# { "$set": { "name": "Minnie" } }).modified_count, "documents updated.")

def deleteOneInCol(dbClient, dbName, dbCollection, queryDict):
    #return selected query.
    print('Deleting the following documents')
    findQueryCol(dbClient, dbName, dbCollection, queryDict)
    dbClient[dbName][dbCollection].delete_one(queryDict)

def deleteManyInCol(dbClient, dbName, dbCollection, queryRegexDict):
    # Expect: { 'key': { "$regex": value } }
    # Ex: { "address": {"$regex": "^S"} }
    #return selected query.
    print('Deleting the following documents')
    findQueryCol(dbClient, dbName, dbCollection, queryRegexDict)
    print(dbClient[dbName][dbCollection].delete_many(queryRegexDict).deleted_count, 
        " documents deleted.")

def deleteAllDocInCol(dbClient, dbName, dbCollection):
    # Delete All Documents in a Collection
    print('Deleting All Documents from ', dbCollection)
    findAllInCol(dbClient, dbName, dbCollection)
    print(dbClient[dbName][dbCollection].delete_many({}).deleted_count, 
        " documents deleted.")

def deleteColInDB(dbClient, dbName, dbCollection):
    # Delete Collection
    print('Deleting the collection', dbCollection)
    dbClient[dbName][dbCollection].drop()
    listCollection(dbClient, dbName)    

def mongoHelp():
    print('''listDatabase(dbClient)
checkDatabase(dbClient, dbName)
listCollection(dbClient, dbName)
checkCollection (dbClient, dbName, dbCollection)
insertOneIntoCol(dbClient, dbName, dbCollection, dictJ)
insertManyIntoCol(dbClient, dbName, dbCollection, listDict)
insertManyIdIntoCol(dbClient, dbName, dbCollection, listDict)
findFirstInCol(dbClient, dbName, dbCollection, listDict)
findAllInCol(dbClient, dbName, dbCollection)
headInCol(dbClient, dbName, dbCollection)
findQueryCol(dbClient, dbName, dbCollection, queryDict)
filterQueryCol(dbClient, dbName, dbCollection, queryDict, filterQuery)
findQueryColAscendSortBy(dbClient, dbName, dbCollection, queryDict, sortKey)
findQueryColDescendSortBy(dbClient, dbName, dbCollection, queryDict, sortKey)
updateOneInCol(dbClient, dbName, dbCollection, queryDict, UpdateDict)
updateManyInColRegex(dbClient, dbName, dbCollection, queryRegexDict, UpdateDict)
deleteOneInCol(dbClient, dbName, dbCollection, queryDict)
deleteManyInCol(dbClient, dbName, dbCollection, queryRegexDict)
deleteAllDocInCol(dbClient, dbName, dbCollection)
deleteColInDB(dbClient, dbName, dbCollection)
mongoHelp()
''')

```
